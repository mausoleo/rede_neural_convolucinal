```python
from PIL import Image, ImageFilter

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('/content/sao_carlos_entrada/sao_carlos_202304.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Definir limites para a floresta (áreas escuras)
limite_floresta_min = 0  # Brilho mínimo para floresta
limite_floresta_max = 100  # Brilho máximo para floresta
mascara_floresta = cv2.inRange(imagem, limite_floresta_min, limite_floresta_max)
```

```python
area_verde = np.sum(mascara_floresta > 0)
area_total = mascara_floresta.shape[0] * mascara_floresta.shape[1]
porcentagem_verde = (area_verde / area_total) * 100
print(f"A porcentagem de área verde é: {porcentagem_verde:.2f}%")
```

```python
# Exibir imagem
nome_arquivo = 'sao_carlos_202304_40.73'
plt.imshow(mascara_floresta, cmap='gray')
plt.axis('off')
plt.savefig(f'/content/sao_carlos_saida/{nome_arquivo}.png', bbox_inches="tight")
plt.show()
```
